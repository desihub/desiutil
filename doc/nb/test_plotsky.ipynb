```python
from desiutil.plots import plot_sky
```

```python
import astropy.table
```

```python
import astropy.units as u
```

```python
import numpy as np
```

```python
table = astropy.table.Table.read('desimodel/data/footprint/desi-tiles.fits')
```

We are going to plot the DESI tiles in the first three years reading the `desi-tiles.fits` file

```python
desi_3_year_mask = np.logical_and(table['PASS']<4,table['IN_DESI']==1)
```

You can use `Shift+tab` to get more information about `plot_sky` once you type `plot_sky` in a code cell or just type `help(plot_sky)`

```python
help(plot_sky)
```

A working example

```python
plot_sky(table['RA'][desi_3_year_mask]*u.degree,table['DEC'][desi_3_year_mask]*u.degree,data=table['PASS'][desi_3_year_mask],label='DESI tiles after 3 years',discrete_colors=True)
```
